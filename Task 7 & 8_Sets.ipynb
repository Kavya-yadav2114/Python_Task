```python
#1: difference() 
a= {"PUBG","FREEFIRE", "COD"}
b= {"OPERA", "CHROME", "COD"}
c= a.difference(b)
print(c)
```

```python
#Symmetric_Difference
a= {"PUBG","FREEFIRE", "COD"}
b= {"OPERA", "CHROME", "COD"}
c= a.symmetric_difference(b)
print(c)
```
